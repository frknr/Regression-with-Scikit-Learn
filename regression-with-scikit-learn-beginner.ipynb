# Linear Regression - Decision Tree Regression - Random Forest Regression

I have created three different regression models using World Happiness Report 2019 data. I set DGP per capita values as indepentend variables and set score of happiness values as dependent variables.

1) **Linear Regression**

```python
import numpy as np # Linear algebra, creating array and other operations
import pandas as pd # Data processing,reading from CSV file
import matplotlib.pyplot as plt # Ploting
```

First of all we need to read data from the (.csv) file. Then I define our dependent and independent variables from data. 

```python
# Reading data from csv 
data = pd.read_csv("../input/world-happiness/2019.csv",sep=",") 

# Determining feature and target samples, getting related data columns
x = data["GDP per capita"].values.reshape(-1,1)
y = data["Score"].values.reshape(-1,1)
```

```python
# Linear Regression
from sklearn.linear_model import LinearRegression

linear_regression_model = LinearRegression()
linear_regression_model.fit(x,y)

# Prediction
x_space = np.linspace(min(x), max(x)).reshape(-1,1)
predicted_y = linear_regression_model.predict(x_space)

# Plotting regression line and scatter
plt.figure(figsize=[10,10])
plt.scatter(x=x, y=y)
plt.plot(x_space, predicted_y, color="red", linewidth=3)
plt.xlabel("GDP per capita")
plt.ylabel("Happiness Score")
plt.show()

# R^2 Score 
print("R^2 Score: {}".format(linear_regression_model.score(x,y)))
```

* Showing Residuals

```python
# Add vlines in order to show residuals
plt.figure(figsize=[10,10])
plt.scatter(x=x, y=y)
plt.plot(x_space, predicted_y, color="red")
plt.xlabel("GDP per capita")
plt.ylabel("Happiness Score")
# Set only x values ("GDP per capita") to predicting
space_for_residual = linear_regression_model.predict(x).reshape(-1,1)
plt.vlines(x, space_for_residual, y, linestyles="--", colors="black", label="Residuals")
plt.legend()
plt.show()
```

2) **Decision Tree Regression**

```python
##Decision Tree Regression##
from sklearn.tree import DecisionTreeRegressor

tree_regression_model = DecisionTreeRegressor()
tree_regression_model.fit(x,y) 

x_space = np.linspace(min(x), max(x)).reshape(-1,1)
predicted_y = tree_regression_model.predict(x_space)

# Plotting 
plt.figure(figsize=[10,10])
plt.scatter(x=x, y=y)
plt.plot(x_space, predicted_y, color="red")
plt.xlabel("GDP per capita")
plt.ylabel("Happiness Score")
plt.show()

# R^2 Score 
print("R^2 Score: {}".format(tree_regression_model.score(x,y)))
```

3) **Random Forest Regression**


```python
##Random Forest Regression##
from sklearn.ensemble import RandomForestRegressor

y_rfr = data["Score"].values

random_forest_model = RandomForestRegressor(n_estimators=150, random_state=42)
random_forest_model.fit(x,y_rfr)

#Prediction
x_space = np.linspace(min(x), max(x)).reshape(-1,1)
predicted_y = random_forest_model.predict(x_space)

#Plotting
plt.figure(figsize=[10,10])
plt.scatter(x=x, y=y_rfr)
plt.xlabel("GDP per capita")
plt.ylabel("Happiness Score")
plt.plot(x_space, predicted_y, color="red")
plt.show()


# R^2 Score 
print("R^2 Score: {}".format(random_forest_model.score(x,y_rfr)))
```
